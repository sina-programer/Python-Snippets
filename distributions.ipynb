```python
from scipy import stats
import seaborn as sns
import numpy as np
```

```python
sns.set_theme()
```

# Probability Distributions

## Random Variation & Charts

### Discrete

```python
p = 0.2
data_bernoulli = stats.bernoulli.rvs(p=p, size=5000)
# data_bernoulli = np.random.binomial(n=1, p=p, size=5000)
print(data_bernoulli)
sns.histplot(data_bernoulli).set_title(f'Bernoulli({p})')
```

```python
n = 100
p = 0.2
data_binom = stats.binom.rvs(n=n, p=p, size=5000)
# data_ninom = np.random.binomial(n=n, p=p, size=5000)
print(data_binom)
sns.histplot(data_binom).set_title(f'Binomial({n}, {p})')
```

```python
data_multinomial = stats.multinomial.rvs(n=100, p=[.2, .3, .5], size=5000)
# data_multinomial = np.random.multinomial(n=100, pvals=[.2, .3, .5], size=5000)
print(data_multinomial)
```

```python
mu = 5
data_poisson = stats.poisson.rvs(mu=mu, size=5000)
# data_poisson = np.random.poisson(lam=mu, size=5000)
print(data_poisson)
sns.histplot(data_poisson).set_title(f'Poisson({mu})')
```

```python
p = 0.5
data_geom = stats.geom.rvs(p=p, size=5000)
# data_geom = np.random.geometric(p=p, size=5000)
print(data_geom)
sns.histplot(data_geom).set_title(f'Geometric({p})')
```

```python
M, n, N = 20, 7, 12
data_hypergeom = stats.hypergeom.rvs(M, n, N, size=5000)
# data_hypergeom = np.random.hypergeometric(M, N, n, size=5000)
print(data_hypergeom)
sns.histplot(data_hypergeom).set_title(f'Hyper Geometric({M}, {n}, {N})')
```

```python
lambda_, n = 1.4, 19
data_boltzmann = stats.boltzmann.rvs(lambda_, n, size=5000)
print(data_boltzmann)
sns.histplot(data_boltzmann).set_title(f'Boltzmann({lambda_}, {n})')
```

```python
lambda_ = 0.51
data_planck = stats.planck.rvs(lambda_, size=5000)
print(data_planck)
sns.histplot(data_planck).set_title(f'Planck({lambda_})')
```

### Continuous 

```python
loc, scale = 0, 1
data_normal = stats.norm.rvs(loc=loc, scale=scale, size=5000)
# data_normal = np.random.normal(loc=loc, scale=scale, size=5000)
print(data_normal)
sns.histplot(data_normal).set_title(f'Normal({loc}, {scale})')
```

```python
df = 5
data_t = stats.t.rvs(df=df, size=5000)
print(data_t)
sns.histplot(data_t).set_title(f'T({df})')
```

```python
df = 5
data_chi2 = stats.chi2.rvs(df=df, size=2000)
print(data_chi2)
sns.histplot(data_chi2).set_title(f'Chi2({df})')
```

```python
a = 20
data_gamma = stats.gamma.rvs(a=a, size=2000)
print(data_gamma)
sns.histplot(data_gamma).set_title(f'Gamma({a})')
```

```python
c = 20
data_fisk = stats.fisk.rvs(c=c, size=2000)
print(data_fisk)
sns.histplot(data_fisk).set_title(f'Fisk({c})')
```

```python
a, b = 0, 10
data_uniform = stats.uniform.rvs(loc=a, scale=b, size=5000)
# data_uniform = np.random.uniform(low=a, high=b, size=5000)
print(data_uniform)
sns.histplot(data_uniform).set_title(f'Uniform({a}, {b})')
```

```python
scale = 3
data_exponential = stats.expon.rvs(scale=scale, size=5000)
# data_exponential = np.random.exponential(scale=scale, size=5000)
print(data_exponential)
sns.histplot(data_exponential).set_title(f'Exponential({scale})')
```

```python
data_wald = stats.wald.rvs(size=2000)
# data_wald = np.random.wald(size=2000)
print(data_wald)
sns.histplot(data_wald).set_title('Wald')
```

## Probability Density/Mass Function

### Continuous

```python
stats.norm.pdf(0)  # default value of loc=0 & scale=1
```

```python
stats.norm.cdf(0)  # cumulative
```

```python
stats.norm.cdf(0, scale=2)  # the chart is symmetric across x=0; so the more scale doesn't affect the CDF
```

```python
print(stats.norm.cdf(0, loc=1))  # the <loc> param moves the chart across the x axis
print(stats.norm.cdf(-1))
```

### Discrete

```python
stats.binom.pmf(20, n=100, p=.2)
```

```python
stats.binom.cdf(20, n=100, p=.2)
```
